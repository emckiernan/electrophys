Next, we can rectify the EMG signal so that we end up with only positive values. This will later make it possible to calculate an envelope and set a threshold for event detection.

```python
# rectify EMG signal
absSignal = np.absolute(dataf)
timeAbs=np.linspace(0, len(absSignal)/sampleRate, num=len(absSignal))

# plot rectified data
plt.figure(figsize=(18,6))
plt.xlabel(r'time (s)')
plt.ylabel(r'voltage ($\mu$V)')
plt.plot(timeEMG,absSignal, 'b')
plt.xlim(0,max(timeEMG))
#plt.ylim(-500,10000)
plt.show()
```

Now we apply a low-pass Butterworth filter with a cutoff of 10 Hz. This will give us the envelope of the rectified EMG signal. 

```python
# low-pass Butterworth filter for envelope detection
lowp = 8
sfreq = sampleRate
low_pass = lowp/sfreq
b, a = sc.signal.butter(4, low_pass, btype='lowpass')
datafrle = filtfilt(b, a, absSignal)
```

```python
# plot rectified EMG signal with superimposed envelope detection with low-pass filter
plt.figure(figsize=(18,6))
plt.xlabel('time (s)')
plt.ylabel('voltage ($\mu$V)')
plt.plot(timeAbs,absSignal, 'b', alpha=0.5)
plt.plot(timeAbs,datafrle*3, 'r', linewidth=2.0) # multiply envelope to see data fit
plt.xlim(0,max(timeEMG))
#plt.ylim(-500,8000)
plt.show()
```

# Analyzing the data

If we plot just the EMG envelope, then we can estimate the best value to use as a threshold to detect muscle contraction on/start and off/stop times. We want a value low enough to include the majority of the contraction, but large enough to exclude noise and small muscle twitches (e.g., see dashed line below). 

Before continuing with analysis, we define a function that will allow us to quickly run the basic data extraction, visualization, and signal processing on EMG recordings. This way we can easily look at different recordings just by changing the file name, or even batch run multiple files.

### Muscle fatigue

As muscles contract, they also fatigue, especially if a significant force is used or if the contraction is prolonged. Fatigue can be seen within a single EMG recording during sustained contraction, or across EMG recordings with repeated contractions. There are several ways we can visualize and quantify fatigue. 

#### Muscle fatigue during a single trial
The following recording was obtained from the bicep muscle while performing a sustained concentric contraction, i.e. holding a weight with the arm bent at 90 degrees for 50 seconds (contraction starts around t=5s and ends at t=55s). 

```python
def analyzeEMG(file,graph):
    # open .wav file by specifying the path and filename
    record = wave.open(file)
    # extract number of channels, sample rate, data
    numChannels = record.getnchannels() # number of channels
    N = record.getnframes() # humber of frames
    sampleRate = record.getframerate() # sampling rate
    # extract data from the .wav file
    dstr = record.readframes(N * numChannels)
    waveData = np.frombuffer(dstr, np.int16)
    # print('The recording has %d channel(s).' % (numChannels))
    # print('The sampling rate of the recording is %d Hz.' % (sampleRate))
    # calculate time window
    timeEMG=np.linspace(0, len(waveData)/sampleRate, num=len(waveData))
    # calculate frequency
    freq = 1/np.mean(np.diff(timeEMG))
    # band-pass Butterworth filter at 10-400Hz
    b, a = butter(2, ([10, 400]/(freq/2)), btype = 'bandpass')
    dataf = filtfilt(b, a, waveData)
    # rectify filtered EMG signal
    absSignal = np.absolute(dataf)
    timeAbs=np.linspace(0, len(absSignal)/sampleRate, num=len(absSignal))
    # low-pass Butterworth filter for envelope detection
    lowp = 8
    sfreq = sampleRate
    low_pass = lowp/sfreq
    b, a = sc.signal.butter(4, low_pass, btype='lowpass')
    datafrle = filtfilt(b, a, absSignal)
    # save relevant data in array
    xx={'absSignal':absSignal,\
        'dataf':dataf,\
        'datafrle':datafrle,\
        'timeAbs':timeAbs,\
        'timeEMG':timeEMG,\
        'waveData':waveData,}   
    # plot rectified EMG signal with superimposed envelope detection 
    if graph:
        plt.figure(figsize=(20,6))
        plt.plot(timeAbs,datafrle*3, 'r', linewidth=2.0) # multiply envelope to see data fit
        plt.plot(timeAbs,absSignal, 'b', alpha=0.5)
        plt.title('Rectified EMG with envelope', fontsize=18)
        plt.ylabel(r'voltage ($\mu$V)')
        plt.xlim(0,max(timeEMG))
        plt.ylim(0,max(absSignal)+1000)
        plt.xlabel('time (s)')
        plt.ylabel(r'voltage ($\mu$V)')
    
    return xx
```

```python
waveData=analyzeEMG(file='/Users/nelion/Documents/pohua/ElectroPhys_FC-UNAM/EMG_dataManuals/mc/mc6.wav',graph=1)
```

```python
def shorten_signal(signal, threshold):
    dif_signal = np.diff(signal)
    startind=np.where(dif_signal>threshold)[0][0]
    
    endind=np.where(dif_signal[::-1]>threshold)[0][0]
    endind=len(dif_signal)-endind
    return(signal[startind:endind])

workdata_abs = shorten_signal(waveData['absSignal'],60)
```

```python
plt.figure(figsize=(20,7))
plt.subplot(211)
ax = plt.gca()
ax.plot(waveData['timeEMG'],waveData['absSignal'],linewidth=.3)
plt.title('Full EMG')
plt.ylabel(r'voltage ($\mu$V)')
plt.xlabel('time (s)')

plt.subplot(212)
ax = plt.gca()
ax.plot(np.arange(0,len(workdata_abs))/sampleRate,workdata_abs,linewidth=.3)
plt.title('Shortened EMG')
plt.ylabel(r'voltage ($\mu$V)')
plt.xlabel('time (s)')

plt.tight_layout()
```

As we can see in the above recording, the amplitude of the signal decreases as the recording progresses and the muscle presumably fatigues. One way to quantify this is to bin the data and find the maximum amplitude of the signal within each bin. For ease, we divide the data into 1 second bins. 

```python
def overlapping_windows(data,npoints,step):

    datapoints = np.size(data)
    anchorpoints = np.arange(0,datapoints-npoints,step)  #Puntos ancla (en los que empieza la ventana de tiempo)
    n_overlap_wind = len(anchorpoints) # Número de ventanas traslapables
  
    ov_data = [data[anchorpoints[i]:anchorpoints[i]+npoints] for i in range(0,n_overlap_wind)]
    
    return(ov_data)
```

```python
windowed_abs_emg = overlapping_windows(workdata_abs,sampleRate,sampleRate)

np.size(pd.DataFrame(windowed_abs_emg[0]))
```

Next, we iterate over the bins, calculating the maximum amplitude for each bins to see how it changes over the course of the recording. 

```python
max_window = list(map(max,windowed_abs_emg))
mean_window = list(map(np.mean,windowed_abs_emg))
var_window = list(map(np.var,windowed_abs_emg))
```

Now we plot the results, with each point representing the maximum amplitude of the EMG signal within each bin.

```python
fig = plt.figure(figsize=(16,8))
ax = plt.subplot(311)
plt.plot(max_window,marker='.',ms=8)
plt.ylabel('Max Amplitude')

ax = plt.subplot(312)
plt.plot(mean_window,marker='.',ms=8)
plt.ylabel('Mean Amplitude')

ax = plt.subplot(313)
plt.plot(var_window,marker='.',ms=8)
plt.ylabel('Amplitude Variance')

plt.tight_layout()
```

The above graph shows that the maximum amplitude decreases slightly as the recording progresses. However, outliers make the trend a less clear than it might be. Instead, we can also iterate over the bins and calculate the mean amplitude to see if that gives us a clearer picture of the decrease due to fatigue.

In this case, the trend is much clearer and we can see how the mean amplitude decreases throughout the recording, indicating muscle fatigue. One crude but effective way to calculate the size of the decrease is by taking the difference between the peak mean amplitudes at the beginning (bin 6) and end (bin 54) of the recording.

# Analysis in the frequency domain (applying Fourier transform)

Muscle fatigue is also marked by a decrease in the frequency of the EMG signal, which can result from both a decrease in the firing rate of individual motor units and a decrease in the total number of active motor units. We can quantify this by iterating over the 1 second bins as before, but this time calculating the median power frequency as follows.

### To analyze the recording in the frequency domain we need to use the signal with the Fourier preprocessing, not the rectified signal:

```python
workdata_f = shorten_signal(waveData['dataf'],60)
windowed_emg_f = overlapping_windows(workdata_f,sampleRate,sampleRate)
```

```python
###########
# https://stackoverflow.com/questions/1523814/
# units-of-a-fourier-transform-fft-when-doing-spectral-analysis-of-a-signal/1524535
###########

def PSD(data):
    ps = np.abs(np.fft.fft(data))**2
    time_step = 1 / sampleRate
    freqs = np.fft.fftfreq(data.size, time_step)
    idx = np.argsort(freqs)
    goodinds = np.where(freqs[idx]>0)[0]
    F = freqs[idx][goodinds]
    P = ps[idx][goodinds]
    return([F,P])

F,P = PSD(windowed_emg_f[0])
```

```python
fig = plt.figure(figsize=(10,12))

ax = plt.subplot(311)
ax.plot(np.arange(0,sampleRate)/sampleRate,windowed_emg_f[0])
#plt.xlim(10,400)
plt.xlabel('Time [s]')
plt.ylabel('Voltage')
plt.title('Bin 0 - Time Domain',fontsize=18)

ax = plt.subplot(312)
ax.plot(F,P)
plt.xlim(10,400)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Voltage^2') 
plt.title('Bin 0 - PSD',fontsize=18)

ax = plt.subplot(313)
ax.plot(F,np.log(P))
plt.xlim(10,400)
plt.xlabel('Frequency (Hz)')
plt.ylabel('log(Voltage^2)') 
plt.title('Bin 0 - log(PSD)',fontsize=17)

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(12,7))

ax = plt.subplot(111)

ax.plot(PSD(windowed_emg_f[0])[0],
        PSD(windowed_emg_f[0])[1],
        label='First Window',alpha=.4
       )


ax.plot(PSD(windowed_emg_f[10])[0],
        PSD(windowed_emg_f[10])[1],
        label='Window 10',alpha=.4
       )

ax.plot(PSD(windowed_emg_f[-1])[0],
        PSD(windowed_emg_f[-1])[1],
        label='Last Window',alpha=.8)
#plt.ylim(20,35)
plt.xlim(10,400)
plt.legend()
plt.xlabel('Frequency (Hz)')
plt.ylabel('log($V^2$)') 
```

# Median Power Frequency

```python
#https://www.intechopen.com/books/computational-intelligence-in-electromyography-analysis-a-perspective-on-current-applications-and-future-challenges/
#     the-usefulness-of-mean-and-median-frequencies-in-electromyography-analysis

def get_PSD(data):
    ps = np.abs(np.fft.fft(data))**2
    time_step = 1 / sampleRate
    #time_step=1
    freqs = np.fft.fftfreq(data.size, time_step)
    idx = np.argsort(freqs)
    goodinds = np.where(freqs[idx]>0)[0]
    F = freqs[idx][goodinds]
    P = ps[idx][goodinds]
    
    return([F,P])


def Median_Power_Freq(data):
    F,P = get_PSD(data)
     
    return(round(sum(P)/2,5))

def Power_Integral(data,lb,ub):
    #lb=0
    #ub=45
    F,P = get_PSD(data)
    Δ = 1/len(F)
    Pow_freqBin = Δ*P 
    return(round((sum(Pow_freqBin[lb:ub])),5))


################################################
MDF = list(map(Median_Power_Freq,windowed_emg_f))
################################################

fig = plt.figure(figsize=(12,4))

ax = plt.subplot(111)
ax.plot(MDF
        ,marker='.',ms=7)
plt.ylabel('Median Power Freq')
plt.xlabel('Time Bin')

plt.tight_layout()
```
