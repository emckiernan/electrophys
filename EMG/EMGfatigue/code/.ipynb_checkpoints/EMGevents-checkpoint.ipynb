# Event Detection
In this notebook, we will introduce a tool that will help us identify pulses in the EMG that are potentially related with the accumulated contribution of individual action potencials. We will introduce a simple method of detection.

```python
def shorten_signal(signal, threshold):
    dif_signal = np.diff(signal)
    startind=np.where(dif_signal>threshold)[0][0]
    
    endind=np.where(dif_signal[::-1]>threshold)[0][0]
    endind=len(dif_signal)-endind
    return(signal[startind:endind])
```

```python
def analyzeEMG(file,graph):
    # open .wav file by specifying the path and filename
    record = wave.open(file)
    # extract number of channels, sample rate, data
    numChannels = record.getnchannels() # number of channels
    N = record.getnframes() # humber of frames
    sampleRate = record.getframerate() # sampling rate
    # extract data from the .wav file
    dstr = record.readframes(N * numChannels)
    waveData = np.frombuffer(dstr, np.int16)
    # print('The recording has %d channel(s).' % (numChannels))
    # print('The sampling rate of the recording is %d Hz.' % (sampleRate))
    # calculate time window
    timeEMG=np.linspace(0, len(waveData)/sampleRate, num=len(waveData))
    # calculate frequency
    freq = 1/np.mean(np.diff(timeEMG))
    # band-pass Butterworth filter at 10-400Hz
    b, a = butter(2, ([10, 400]/(freq/2)), btype = 'bandpass')
    dataf = filtfilt(b, a, waveData)
    # rectify filtered EMG signal
    absSignal = np.absolute(dataf)
    timeAbs=np.linspace(0, len(absSignal)/sampleRate, num=len(absSignal))
    # low-pass Butterworth filter for envelope detection
    lowp = 8
    sfreq = sampleRate
    low_pass = lowp/sfreq
    b, a = sc.signal.butter(4, low_pass, btype='lowpass')
    datafrle = filtfilt(b, a, absSignal)
    # save relevant data in array
    xx={'absSignal':absSignal,\
        'dataf':dataf,\
        'datafrle':datafrle,\
        'timeAbs':timeAbs,\
        'timeEMG':timeEMG,\
        'waveData':waveData,}   
    # plot rectified EMG signal with superimposed envelope detection 
    if graph:
        plt.figure(figsize=(20,6))
        plt.plot(timeAbs,datafrle*3, 'r', linewidth=2.0) # multiply envelope to see data fit
        plt.plot(timeAbs,absSignal, 'b', alpha=0.5)
        plt.title('Rectified EMG with envelope', fontsize=18)
        plt.ylabel(r'voltage ($\mu$V)')
        plt.xlim(0,max(timeEMG))
        plt.ylim(0,max(absSignal)+1000)
        plt.xlabel('time (s)')
        plt.ylabel(r'voltage ($\mu$V)')
    
    return xx
```

For the detection of pulses, we will use a method from https://mcspydatatools.readthedocs.io/en/latest/McsPy-Tutorial_DataAnalysis.html


Let $Y(t)$ be the amplitude of the record at time $t$.

Using this method, $Y(t)$ is considered a pulse (spike) if it follows the next conditions:

1) $Y(t) > threshold$.

2) $\tau > dead time$, where $\tau$ is the time between two consecutive threshold crossings.  

```python
def detect_threshold_crossings(signal, fs, threshold, dead_time):
    """
    Detect threshold crossings in a signal with dead time and return them as an array

    The signal transitions from a sample above the threshold to a sample below the threshold 
    for a detection and the last detection has to be more than dead_time apart from the 
    current one.

    :param signal: The signal as a 1-dimensional numpy array
    :param fs: The sampling frequency in Hz
    :param threshold: The threshold for the signal
    :param dead_time: The dead time in seconds.
    """
    dead_time_idx = dead_time * fs #Cantidad de muestras durante el tiempo muerto
    
    threshold_crossings = np.diff((signal <= threshold).astype(int) > 0).nonzero()[0]
    
    distance_sufficient = np.insert(np.diff(threshold_crossings) >= dead_time_idx, 0, True)
    
    while not np.all(distance_sufficient):
         # repeatedly remove all threshold crossings that violate the dead_time
        threshold_crossings = threshold_crossings[distance_sufficient]
        distance_sufficient = np.insert(np.diff(threshold_crossings) >= dead_time_idx, 0, True)
    return threshold_crossings

def get_next_minimum(signal, index, max_samples_to_search):
    """
    Returns the index of the next minimum in the signal after an index

    :param signal: The signal as a 1-dimensional numpy array
    :param index: The scalar index
    :param max_samples_to_search: The number of samples to search for a minimum after the index
    """
    search_end_idx = min(index + max_samples_to_search, signal.shape[0])
    min_idx = np.argmin(signal[index:search_end_idx])
    return index + min_idx

def align_to_minimum(signal, fs, threshold_crossings, search_range):
    """
    Returns the index of the next negative spike peak for all threshold crossings

    :param signal: The signal as a 1-dimensional numpy array
    :param fs: The sampling frequency in Hz
    :param threshold_crossings: The array of indices where the signal crossed the detection threshold
    :param search_range: The maximum duration in seconds to search for the minimum after each crossing
    """
    search_end = int(search_range*fs)
    aligned_spikes = [get_next_minimum(signal, t, search_end) for t in threshold_crossings]
    return np.array(aligned_spikes)
```

# Setting the event detector...

```python
threshold=13000
dt = .002 # dead time of 2 ms
tm = .001 # search range of 1 ms

crossings = detect_threshold_crossings(xx['absSignal'], fs, threshold, dt) 
spks = align_to_minimum(xx['absSignal'], fs, crossings, tm) 
```

```python
spike_time =spks/sampleRate
cross_time =crossings/sampleRate
    
plt.figure(figsize=(20,8))
plt.title('EMG')
plt.xlabel('Time [s]')
plt.ylabel('Voltage ($\mu$V)')
plt.plot(xx['timeAbs'],xx['absSignal'],linewidth=.1)
plt.xlim(0,max(xx['timeEMG']))
plt.axhline(y=threshold, color='k',linewidth=3.0, linestyle='--') 

for n in range(0, len(spike_time)):
    plt.vlines(x=spike_time[n], ymin=xx['absSignal'].max() *1.01, 
                            ymax=xx['absSignal'].max() *1.05,
                            color='k',linewidth=1.0, linestyle='-')
plt.show()
```

# Counting the number spike detections per bin
Now we will count the number of spikes as the time evolves. As a result of fatigue, we should expect a decrease in the number of pulses with time.

```python
bin_size = 5 #Bin length (in seconds)

spikeBins = list(map(int,spike_time/bin_size)) #Bins that contain spikes
spikeBins = pd.DataFrame(spikeBins,columns=['Bin_Index'])

SpikeBin_count = pd.value_counts(spikeBins['Bin_Index']) #Count the number of spikes
                                                         #per bin
                                                         
SpikeBin_count = pd.DataFrame(SpikeBin_count)  #transform that to a DataFrame
SpikeBin_count.columns = ['Spike_Count']
SpikeBin_count
```

```python
numBins = int(len(xx['timeAbs'])/sampleRate/bin_size) #Number of bins

all_counts = np.zeros(numBins) #Initialize array
all_counts[SpikeBin_count.index] = SpikeBin_count['Spike_Count']
all_counts = pd.DataFrame(all_counts,columns=['Spike_Count'])


#Plotting the result:
from matplotlib.ticker import MaxNLocator
ax = plt.figure().gca()
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
all_counts['Spike_Count'].plot(kind='bar')
plt.xlabel('Bin')
plt.title('Conteo de Spikes por Bin de {} segundos'.format(bin_size))
plt.tight_layout()
```

# Finallly we set up an interactive enviroment in order to explore the event detection with different parameters:

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def plt_event(data,thresh,bin_size):
    
    fs=sampleRate
    dead_time=.002
    min_time=.001
    crossings = detect_threshold_crossings(data, fs, thresh, dead_time) # dead time of 2 ms
    spks = align_to_minimum(data, fs, crossings, min_time) # search range 1 ms
    time_data = np.arange(0,len(data))/sampleRate
    
    spike_time =spks/sampleRate
    cross_time =crossings/sampleRate

    print('{} Spikes detectados en total'.format(len(spike_time)))

    plt.figure(figsize=(20,8))
    
    #######################
    #FIGURE 1: Event Raster
    #######################

    plt.xlabel('Time [s]')
    plt.ylabel('Voltage ($\mu$V)')
    plt.plot(time_data,data,linewidth=.1)
    plt.axhline(y=threshold, color='k',linewidth=3.0, linestyle='--') 

    for n in range(0, len(spike_time)):
        plt.vlines(x=spike_time[n], ymin=data.max() *1.01, 
                                ymax=data.max() *1.05,
                                color='k',linewidth=1.0, linestyle='-')
    plt.show()
    ##########################
    #FIGURE 2: Events per Bin
    ##########################
 
    plt.subplot(2,1,2)
    ax = plt.gca()
    
    spikeBins = list(map(int,spike_time/bin_size)) #índices de bins con spikes
    spikeBins = pd.DataFrame(spikeBins,columns=['Bin_Index'])

    SpikeBin_count = pd.value_counts(spikeBins['Bin_Index']) #Cuenta el número de apariciones 
                                                             # de cada índice (númerp de spikes por bin)
    SpikeBin_count = pd.DataFrame(SpikeBin_count)
    SpikeBin_count.columns = ['Spike_Count']
    numBins = int(len(data)/sampleRate/bin_size)

    all_counts = np.zeros(numBins)
    all_counts[SpikeBin_count.index] = SpikeBin_count['Spike_Count']
    all_counts = pd.DataFrame(all_counts,columns=['Spike_Count'])

    ax.yaxis.set_major_locator(MaxNLocator(integer=True))
    all_counts['Spike_Count'].plot(kind='bar',width=.85)
    plt.xlabel('Bin')
    plt.title('Conteo de Spikes por Bin de {} segundos'.format(bin_size))
    
    ##########################
    #DISPLAY GRAPHS
    ##########################
    plt.tight_layout()
    plt.show()
```
