```python
import matplotlib.pyplot as plt 
import numpy as np

def espiro (archivo):
    # open file and then strip out extra characters
    f = open(archivo,'r')
    data = f.readlines()
    #print data
    #data = [x.strip() for x in data]

    # replace commas with decimals points throughout list
    dataNew = list()

    for i in range(0, len(data)):
        xx = data[i].replace("," , ".")
        dataNew.append(xx)
    #print dataNew

    # read through the new data and convert to floats
    begin = 1
    separator = '\t'

    # set up empty lists for x and y values
    xvals = list()
    yvals1 = list()
    #yvals2 = list()

    # split data values in the list, convert to floats, and save as x and y values
    for i in range(0, len(dataNew)-begin):
        line = dataNew[i]
        #print line
        datos = line.split(separator)
        #print datos
        xx = float(datos[0])
        y1 = float(datos[1])
        #y2 = float(datos[2])
        #print xx,y1
        xvals.append(xx)
        yvals1.append(y1)
        #yvals2.append(y2)

    # plot the data
    plt.plot(xvals,yvals1)
    plt.title(archivo)
    plt.xlabel('tiempo (s)')
    plt.ylabel('volumen caudal (L/s)')
    #plt.set_aspect('equal',adjustable='box') #Guarda la proporcion entre los ejes x y y.
    plt.show()
```

```python
espiro("S1_4.txt")
espiro("S1_5.txt")
espiro("S2_11.txt")
espiro("S3_1.txt")
espiro("S3_2.txt")
```
