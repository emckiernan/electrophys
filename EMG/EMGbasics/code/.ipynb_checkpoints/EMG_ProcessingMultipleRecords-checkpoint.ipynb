## Muscle fatigue

As muscles contract, they also fatigue, especially if a significant force is used or if the contraction is prolonged. Fatigue can be seen within a single EMG recording during sustained contraction, or across EMG recordings with repeated contractions. There are several ways we can visualize and quantify fatigue.

```python
def analyzeEMG(file,graph):
    # open .wav file by specifying the path and filename
    record = wave.open(file)
    # extract number of channels, sample rate, data
    numChannels = record.getnchannels() # number of channels
    N = record.getnframes() # humber of frames
    sampleRate = record.getframerate() # sampling rate
    # extract data from the .wav file
    dstr = record.readframes(N * numChannels)
    waveData = np.frombuffer(dstr, np.int16)
    # print('The recording has %d channel(s).' % (numChannels))
    # print('The sampling rate of the recording is %d Hz.' % (sampleRate))
    # calculate time window
    timeEMG=np.linspace(0, len(waveData)/sampleRate, num=len(waveData))
    # calculate frequency
    freq = 1/np.mean(np.diff(timeEMG))
    # band-pass Butterworth filter at 10-400Hz
    b, a = butter(2, ([10, 400]/(freq/2)), btype = 'bandpass')
    dataf = filtfilt(b, a, waveData)
    # rectify filtered EMG signal
    absSignal = np.absolute(dataf)
    timeAbs=np.linspace(0, len(absSignal)/sampleRate, num=len(absSignal))
    # low-pass Butterworth filter for envelope detection
    lowp = 8
    sfreq = sampleRate
    low_pass = lowp/sfreq
    b, a = sc.signal.butter(4, low_pass, btype='lowpass')
    datafrle = filtfilt(b, a, absSignal)
    # save relevant data in array
    xx={'absSignal':absSignal,\
        'dataf':dataf,\
        'datafrle':datafrle,\
        'timeAbs':timeAbs,\
        'timeEMG':timeEMG,\
        'waveData':waveData,}   
    # plot rectified EMG signal with superimposed envelope detection 
    if graph:
        plt.figure(figsize=(20,6))
        plt.plot(timeAbs,datafrle*3, 'r', linewidth=2.0) # multiply envelope to see data fit
        plt.plot(timeAbs,absSignal, 'b', alpha=0.5)
        plt.title('Rectified EMG with envelope', fontsize=18)
        plt.ylabel(r'voltage ($\mu$V)')
        plt.xlim(0,max(timeEMG))
        plt.ylim(0,max(absSignal)+1000)
        plt.xlabel('time (s)')
        plt.ylabel(r'voltage ($\mu$V)')
    
    return xx
```

As we can see in the above recording, the amplitude of the signal decreases as the recording progresses and the muscle presumably fatigues. One way to quantify this is to bin the data and find the maximum amplitude of the signal within each bin. For ease, we divide the data into 1 second bins. 

```python
workdata = waveData['dataf']
```

```python
def shorten_signal(signal, threshold):
    dif_signal = np.diff(signal)
    startind=np.where(dif_signal>threshold)[0][0]
    
    endind=np.where(dif_signal[::-1]>threshold)[0][0]
    endind=len(dif_signal)-endind
    return(signal[startind:endind])
#shorten_signal(workdata,50)
```

```python
def overlapping_windows(data,npoints,step):

    datapoints = np.size(data)
    anchorpoints = np.arange(0,datapoints-npoints,step)  #Puntos ancla (en los que empieza la ventana de tiempo)
    n_overlap_wind = len(anchorpoints) # Número de ventanas traslapables
  
    ov_data = [data[anchorpoints[i]:anchorpoints[i]+npoints] for i in range(0,n_overlap_wind)]
    
    return(ov_data)

```

## Functions for the analysis in the frequency domain

```python
#https://www.intechopen.com/books/computational-intelligence-in-electromyography-analysis-a-perspective-on-current-applications-and-future-challenges/
#     the-usefulness-of-mean-and-median-frequencies-in-electromyography-analysis

def get_PSD(data):
    ps = np.abs(np.fft.fft(data))**2
    time_step = 1 / sampleRate
    #time_step=1
    freqs = np.fft.fftfreq(data.size, time_step)
    idx = np.argsort(freqs)
    goodinds = np.where(freqs[idx]>0)[0]
    F = freqs[idx][goodinds]
    P = ps[idx][goodinds]
    
    return([F,P])
    
def Mean_Power_Freq(data):
    F,P = get_PSD(data)
    sum_prod = sum([F[i]*P[i] for i in range(0,len(F))])
    sumPower = sum(P)
    promPower = sum_prod/sumPower
    
    return(round(promPower,5))

def Median_Power_Freq(data):
    F,P = get_PSD(data)
    medianPower = sum(P)/2
     
    return(round(medianPower,5))
```

### Las siguientes funciones actúan directamente sobre el archivo y esto, además de ahorrar memoria, nos permitirá analizar varios archivos de manera rápida.

```python
def get_MDF(path):
    dataf = analyzeEMG(path,0)['dataf']
    shorten_signal(dataf, 50)
    ov_data = overlapping_windows(dataf,sampleRate,sampleRate)
    MDF = list(map(Median_Power_Freq,ov_data))
    
    return(MDF)
```

#### Muscle fatigue across trials

Muscles also fatigue with repeated contractions. We asked subjects to repeat the above exercise (holding a weight with the arm bent at 90 degrees for 50 seconds) for 10 trials with 60 seconds rest between each trial. Trial 0, analyzed above, is the first trial (i.e. before fatigue through repetition) and trial 9 is the last repetition, i.e. after fatigue.

There are several measures we could use to quantify fatigue. One is the **maximum amplitude** of the EMG signal. If repeated trials fatigue the muscle, then the maximum contraction subjects generate during trial 9 should be less than that generated during trial 0, and we will see a decrease in the maximum amplitude. 

```python
import os
```

Next, we will define a function that opens different .wav files and applies the function that computes MDF in time bins.

```python
def loop_MDF(Directory):
    trialMDF=[]
    #Directory = '/Users/Mitsui Myrna/Documents/muscles'
    for filename in os.listdir(Directory):
        if filename.endswith(".wav"): 
            MDF = get_MDF(os.path.join(Directory, filename))
            trialMDF.append(MDF)
    return(trialMDF)  
```

```python
trialMDF = loop_MDF('/Users/nelion/Documents/POHUA/ElectroPhys_FC-UNAM/EMG_dataManuals/mc/')
```

We can see that the objet returned by *loop_MDF(Directory)* is a list that contains the MDF time series for the different trials of the same subject:

```python
plt.figure(figsize=(11,6))
trialMDF = loop_MDF('/Users/nelion/Documents/POHUA/ElectroPhys_FC-UNAM/EMG_dataManuals/mc/')
for i in range(0,len(trialMDF)):
    plt.plot(trialMDF[i],c='Blue',alpha=(i+1)*.35)
    plt.xlabel('time (s)')
    plt.ylabel(r'voltage ($\mu$V)')
```

The colors in the graph go from the lightest to the strongest, with the first trial being the lightest and so on respectively. 

We can see that for all the measures, there is a noticeable decrease when we compare the first to the last trial, indicating that the protocol of repeated contractions induced fatigue in the bicep muscle. Next we made the graphs of all the trials individually.



```python
plt.figure(figsize=(13,10))
for i in range(0,len(trialMDF)):
    ax=plt.gca()
    plt.subplot(len(trialMDF),1,i+1)
    plt.plot(trialMDF[i],c='blue')
    
    plt.ylim(0,max(trialMDF[0])*1.1)
    plt.title('Trial: %i' % i)
    plt.xlabel('time (s)')
    plt.ylabel(r'voltage ($\mu$V)');
    
    plt.tight_layout()
```

### Mean
To continue exploring all the trials of the same individual, the mean was taken in the different time windows and individually plotted

```python
def get_mean(path):
    dataf = analyzeEMG(path,0)['absSignal']
    shorten_signal(dataf, 50)
    ov_data = overlapping_windows(dataf,sampleRate,sampleRate)
    mean = list(map(np.mean,ov_data))
    return(mean)
    
```

```python
def loop_mean(Directory):
    trialmean=[]
    #Directory = '/Users/Mitsui Myrna/Documents/muscles'
    for filename in os.listdir(Directory):
        if filename.endswith(".wav"): 
            mean = get_mean(os.path.join(Directory, filename))
            trialmean.append(mean)
    return(trialmean)  
```

## Spectral Analysis along different trials

After noticing that fatigue might be related to a decrease in all the metrics we can also expect to find changes in the Fourier Coefficients associated to each frequency (which are related to the energy contributed by each frequency):

```python
wavetrial0=analyzeEMG(file='/Users/nelion/Documents/POHUA/ElectroPhys_FC-UNAM/EMG_dataManuals/mc/mc0.wav',graph=0)
wavetrial6=analyzeEMG(file='/Users/nelion/Documents/POHUA/ElectroPhys_FC-UNAM/EMG_dataManuals/mc/mc6.wav',graph=0)



fig = plt.figure(figsize=(18,6),dpi=100)

#####
#Primer Espectograma
#####
data = shorten_signal(wavetrial0['absSignal'],0)
f, t, sxx1 = sc.signal.spectrogram(data,
                                  sampleRate,
                                  nperseg=sampleRate, 
                                  scaling='spectrum')
f = f[:401]
sxx1 = sxx1[:401]
min_sxx1 = np.min(np.log(sxx1))
max_sxx1 = np.max(np.log(sxx1))

plt.subplot(211)
plt.pcolormesh(np.log(sxx1),cmap='jet',vmin=min_sxx1,vmax=max_sxx1)
plt.ylabel('Frequency [Hz]')
plt.title('Spectogram - Rectified EMG TRIAL_0')

########################################################

#####
#Segundo Espectograma
#####
data = shorten_signal(wavetrial6['absSignal'],0)
f, t, sxx = sc.signal.spectrogram(data,
                                  sampleRate,
                                  nperseg=sampleRate, 
                                  scaling='spectrum')
f = f[:401]
sxx = sxx[:401]
plt.subplot(212)
plt.pcolormesh(np.log(sxx),cmap='jet',vmin=min_sxx1,vmax=max_sxx1)
plt.ylabel('Frequency [Hz]')
plt.title('Spectogram - Rectified EMG TRIAL_6')

plt.xlabel('Time[s]')
#####
#Colorbar
#####
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([1.01, .1, 0.035, 0.8])
plt.colorbar(label='log($Voltage^2$)',cax=cbar_ax,)

plt.draw()
plt.tight_layout()
```

It is not difficult to notice that the first spectogram reaches higher values  associated with the power content per frequency (orange area). This can be observed along the whole frequency band (0-400 Hz).


The second spectogram shows a generalized decrease in the power content contributed by frequency. It is possibble to quickly compare these graphs by noticing the difference in the overall colors of the spectogram, with the first spectogram showing tones closer to red (higher power values), while the second one displays an overall yellow color (lower $power$ values)
